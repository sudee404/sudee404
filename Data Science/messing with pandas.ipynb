```python
#narrowing down to england
de=df[(df["State"]=="England")]
de=de.drop("State",axis=1)
de.head()
```

```python
dp=de[["Date","Product","Unit_Cost","Order_Quantity","Unit_Price"]]
dp=dp.drop_duplicates("Product")
dp=dp.sort_index()
dp.head(20)
```

```python
#sales above 100-great sales
gs=dp[(dp["Unit_Price"]>3000)]
gs.set_index("Date",inplace=True)
gs.head(12)
```
